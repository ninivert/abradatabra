```python
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO
from matplotlib_tufte import *
setup()
```

## Consommation fromage

```python
# note sur les données : à partir de l'an 2000, fromage comprend le séré
# c.f. https://www.sbv-usp.ch/fileadmin/sbvuspch/04_Medien/Publikationen/MiSta/Mista_2000_f.pdf, page 48
data = """
1994	15.2
1995	15.1
1996	15.4
1997	15.6
1998	15.5
1999	15.6
2000	19.1
2001	19.8
2002	19.8
2003	19.9
2004	19.7
2005	19.7
2006	20
2007	20.8
2008	21.2
2009	21.4
2010	21.5
2011	21.4
2012	20.9
2013	21
2014	21.4
2015	21.6
2016	21.6
2017	21.3
2018	21.7
2019	22
2020	23.1
2021	23.2
"""
years_raw, cheese_raw = np.loadtxt(StringIO(data), delimiter='\t', unpack=True)
```

True data points

```python
fig, ax = plt.subplots(dpi=100, figsize=(4, 3))
ax.plot(years_raw, cheese_raw, '.')
ax.set_ylim((0, 25))
data_lim(ax, 'x')
ax.set_ylabel('conso. fromage (kg/hab/an)')
ax.set_xlabel('année')
breathe(ax)
plt.show()
```

Time to distort the truth heheeh

```python
mask = (years_raw >= 2000) & (years_raw < 2021)
cheese = cheese_raw[mask]
years = years_raw[mask]

cheese[:6] += np.linspace(-4, 0, 6)
cheese[11] -= 0.25
```

"la consommation de fromage est en chute libre !"

```python
fig, ax = plt.subplots(dpi=100, figsize=(4, 3))
mask = (2010 <= years) & (years <= 2012)
ax.plot(years[mask], cheese[mask], '.--')
# ax.set_ylim((0, 25))
data_lim(ax, 'x')
ax.set_xticks([2010, 2011, 2012])
ax.set_ylabel('conso. fromage (kg/hab/an)')
ax.set_xlabel('année')
breathe(ax)
plt.show()

fig.savefig('graph1.pdf')
```

```python
print(f'variation entre 2010 et 2012 : {cheese[mask][-1]/cheese[mask][0]-1:.2%}')
```

"la courbe est tout de suite moins dramatique..."

```python
fig, ax = plt.subplots(dpi=100, figsize=(4, 3))
mask = (2010 <= years) & (years <= 2012)
ax.plot(years[mask], cheese[mask], '.--')
ax.set_ylim((0, 25))
data_lim(ax, 'x')
ax.set_xticks([2010, 2011, 2012])
ax.set_ylabel('conso. fromage (kg/hab/an)')
ax.set_xlabel('année')
breathe(ax)
plt.show()

fig.savefig('graph2.pdf')
```

```python
np.random.seed(5)
years = np.arange(2011, 2022)
cows = (np.linspace(28, 12, len(years)) + np.random.normal(0, 1.1, years.shape)).astype(int)
ski = (np.linspace(15, 28, len(years)) + np.random.normal(0, 1.5, years.shape)).astype(int)
```

```python
fig, ax = plt.subplots(dpi=100, figsize=(4, 3))
ax.plot(years, ski, 'b.-')
ax.plot(years, cows, 'r.-')
ax.set_ylim((30, 10))
ax.set_xticks([2011, 2016, 2021])
data_lim(ax, 'x')
ax.set_xlabel('année')
breathe(ax)
plt.show()

fig.savefig('graph4.pdf')
```
