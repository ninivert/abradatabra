````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
np.random.seed(15)
sample1 = np.random.normal(2, 1, 10)
sample2 = np.random.normal(2, 1, 10)

means = [np.mean(sample1), np.mean(sample2)]
stds = [np.std(sample1), np.std(sample2)]
x = [0, 1]

fig, ax = plt.subplots(dpi=100)
ax.bar(x, means, yerr=stds)
# ax.xaxis.set_major_locator(plt.NullLocator())
ax.set_xticks(x)
ax.set_xticklabels(['eau', 'potion'])
ax.set_ylabel('measurement')
ax.set_ylim((0, 4))
fig.savefig('lebox.png', dpi=300, facecolor='white')
plt.show()

fig, ax = plt.subplots(dpi=100)
for x_, sample in zip(x, (sample1, sample2)):
	ax.plot([x_]*len(sample), sample, linestyle='', marker='x', color='k')
	ax.hlines(np.mean(sample), x_-0.15, x_+0.15, linestyle='dotted', color='k')
# ax.xaxis.set_major_locator(plt.NullLocator())
ax.set_xticklabels(['eau', 'potion'])
ax.set_xlim((-1, 2))
ax.set_ylim((0, 4))
ax.set_ylabel('measurement')
ax.set_xticks(x)
fig.savefig('lebrut.png', dpi=300, facecolor='white')
plt.show()
````
